```python
code_states = code_sheet[code_sheet['Summary Level'] == 40]
code_states.head() 
```

```python
code_states.tail()
```

```python
crime_data = pd.read_csv("Resources/nanda_crime_county_2002-2014_01P.csv")
```

```python
crime_data.head()
```

```python
crime_data.columns
```

```python
crime_data_renamed = crime_data.rename(columns={'FIPS_ST': 'State Code (FIPS)',
                                                'FIPS_CTY': 'County Code (FIPS)',
                                                'YEAR': 'year',
                                                'MURDER': 'murder',
                                                'RAPE': 'rape',
                                                'ROBBERY': 'robbery',
                                                'AGASSLT': 'aggravated_assaults',
                                                'BURGLRY': 'burglary',
                                                'LARCENY': 'larceny',
                                                'MVTHEFT': 'auto_theft',
                                                'ARSON': 'arson'})
                                                
```

```python
crime_data_small = crime_data_renamed.drop(columns=['STCOFIPS', 'STUDYNO', 'EDITION', 'PART', 'IDNO', 'CPOPARST',
                                            'CPOPCRIM', 'AG_ARRST', 'AG_OFF', 'COVIND', 'INDEX', 'MODINDX',
                                            'VIOL', 'PROPERTY'])
```

```python
crime_data_small.head()
```

```python
state_merge = pd.merge(code_states, crime_data_small, on = 'State Code (FIPS)', how = 'inner')
state_merge.head()
```

```python
state_merge.to_csv('Resources/merge_state.csv')
```

```python
state_merge.shape
```

```python
state_merge.dropna()
```

```python
state_merge = state_merge.rename(columns = {'County Code (FIPS)_y': 'County Code (FIPS)'})
state_merge.head()
```

```python
state_county_merge = pd.merge(code_sheet, state_merge, on = 'County Code (FIPS)', how = 'inner')
state_county_merge.head()
```

```python
state_county_merge.tail()
```

```python
state_county_merge.dropna()
```

```python
crime_df_final = state_county_merge.rename(columns = {'Area Name (including legal/statistical area description)_x': 'County',
                                                     'Area Name (including legal/statistical area description)_y': 'State'})
```

```python
only_county = crime_df_final[crime_df_final['Summary Level_x'] != 61]
only_county.shape
```

```python
only_county = only_county[only_county['Summary Level_x'] != 162]
only_county.shape
```

```python
only_county = only_county[only_county['Summary Level_x'] != 170]
only_county.shape
```

```python
only_county.tail()
```

```python
crime_df_final = only_county.drop(columns=['Summary Level_x', 'State Code (FIPS)_x', 'County Code (FIPS)', 'County Subdivision Code (FIPS)_x',
                                              'Place Code (FIPS)_x', 'Consolidtated City Code (FIPS)_x', 'Summary Level_y', 'State Code (FIPS)_y',
                                              'County Code (FIPS)_x', 'County Subdivision Code (FIPS)_y', 'Place Code (FIPS)_y', 'Consolidtated City Code (FIPS)_y'])
crime_df_final
```

```python
crime_df_final.to_parquet('Resources/crime_df_final.parquet')
```
