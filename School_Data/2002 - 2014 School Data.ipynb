```python
import pandas as pd
```

```python
#load and read file
file_path = file_path = "C:\\Users\\LabUser\\Project_3\\School_Data\\ELSI_csv_export.csv"

# Skip to actual data
data = pd.read_csv(file_path, skiprows=6)
data.head()
```

```python
# replace special characters
data_cleaned = data.replace({"†": pd.NA, "–": pd.NA, "‡": pd.NA})
data_cleaned.head()
```

```python
data_cleaned = data_cleaned.apply(pd.to_numeric, errors='ignore')
data_cleaned.head()
```

```python
# drop rows with missing values
data_cleaned = data_cleaned.dropna()
data_cleaned
```
